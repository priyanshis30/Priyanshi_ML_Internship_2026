```python
#Factorial of a number
num=int(input("Enter a number:",))
fact=1

for i in range(1,num+1):
  fact=fact*i
print("Factorial of",num,"is",fact)
```

```python
#Find largest among 3 numbers
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
num3=int(input("Enter third number:"))

if num1>num2 and num1>num3:
  print(num1,"is the largest number")
elif num2>num1 and num2>num3:
  print(num2,"is the largest number")
elif num3>num1 and num3>num2:
  print(num3,"is the largest number")
else:
  print("Enter different numbers")
```

```python
name = input("Enter your name: ")
enroll_no = int(input("Enter your roll number: "))

subjects = ["Maths", "Physics", "Chemistry", "English", "Python", "Physical Education"]

total = 0

for subject in subjects:
    marks = float(input(f"Enter marks in {subject}: "))
    total += marks

percentage = total / len(subjects)

print("Name:", name)
print("Enrollment No:", enroll_no)
print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
```
